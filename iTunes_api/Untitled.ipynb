```python
base_site = "https://itunes.apple.com/search"
```

```python
import requests
url = base_site + "?term=the+beatles&country=us"
```

```python
requests.get(url)
```

```python
r = requests.get(base_site, params = {
    "term" : "the beatles", "country" : "us"
})
r.status_code
```

```python
info = r.json()
import json
print(json.dumps(info, indent = 4))
```

```python
info.keys() #this function return the overall key in the output
```

```python
print(json.dumps(info['results'][0], indent=4))
```

```python
info["resultCount"]
```

```python
r = requests.get(base_site , params = {"term" : "the beatles", "country":"us", "limit" : 200})
r.ok
```

```python
r.json()["resultCount"]
```
